```python


train_dir = "/content/drive/MyDrive/datasets/DATASET/dataset/DATASET/TRAIN"
test_dir  = "/content/drive/MyDrive/datasets/DATASET/dataset/DATASET/TEST"

aug_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    brightness_range=[0.5, 1.5],
    horizontal_flip=True,
    validation_split=0.2
)

train = aug_gen.flow_from_directory(
    train_dir,
    target_size=(150,150),
    batch_size=32,
    subset='training',
    class_mode='binary'
)

val = aug_gen.flow_from_directory(
    train_dir,
    target_size=(150,150),
    batch_size=32,
    subset='validation',
    class_mode='binary'
)
```
